## SMOTE(Synthetic Minority Oversampling Technique)
SMOTE (Synthetic Minority Over-sampling Technique) is a technique used in machine learning to address imbalanced datasets where the minority class has significantly fewer instances than the majority class. SMOTE involves generating synthetic instances of the minority class by interpolating between existing instances.

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv("creditcard.csv")
df.head()
```

```python
df.Class.value_counts()
```

```python
class_0=df[df["Class"]==0]
class_1=df[df["Class"]==1]
```

```python
sns.scatterplot(df["Class"])
```

```python
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
plt.scatter(df['V3'],df['V4'],c=df['Class'])
```

```python
## transform the dataset
oversample=SMOTE()
X,y=oversample.fit_resample(df[['V3','V4']],df['Class'])
```

```python
len(y[y==0])
len(y[y==1])
```

```python
df1=pd.DataFrame(X,columns=['V3','V4'])
df2=pd.DataFrame(y,columns=['Class'])
oversample_df=pd.concat([df1,df2],axis=1)
```

```python
plt.scatter(oversample_df['V3'],oversample_df['V4'],c=oversample_df['Class'])
```
